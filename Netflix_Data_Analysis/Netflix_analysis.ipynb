# Categorizing Vote_Average column

We would cut the Vote_Average values and make 4 categories Popular, Average, Below_Average and Not_Popular to describe it more using categorize_col() function provided above.

```python
def categorize_col(df, col, levels):

    edges = [df[col].describe()['min'],
            df[col].describe()['25%'],
            df[col].describe()['50%'],
            df[col].describe()['75%'],
            df[col].describe()['max']]

    df[col] = pd.cut(df[col], edges, labels = labels, duplicates = 'drop')
    return df
```

```python
labels = ['Not_Popular', 'Below_Average', 'Average', 'Popular']

categorize_col(df, 'Vote_Average', labels)

df['Vote_Average'].unique()
```

# Top 10 Movies by Popularity

```python
top_pop = df.sort_values(by='Popularity', ascending=False)

top_pop_unique = top_pop.drop_duplicates(subset='Title').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='Popularity', y='Title', data=top_pop_unique)

plt.title("Top 10 Unique Movies by Popularity")
plt.tight_layout()
plt.show()
```
